4) Identifiez les 10 principaux genres en fonction de leur popularité et tracez les tendances de diverses caractéristiques sonores (valence, énergie, dansabilité, acoustique) pour ces genres à l'aide d'un graphique à barres groupées : px.bar(top10_genres, x='genres', y=['valence', 'energy', 'danceability', 'acousticness'], barmode='group', title='Tendance de diverses caractéristiques sonores sur les 10 principaux genres').

```python
import pandas as pd
import plotly.express as px
# Calculer la popularité moyenne par genre
genre_popularity = genre_data.groupby('genres')['popularity'].mean().reset_index()
```

```python
# Identifier les 10 genres les plus populaires
top10_genres = genre_popularity.nlargest(10, 'popularity')
# Extraire les caractéristiques sonores pour ces genres
sound_features = ['valence', 'energy', 'danceability', 'acousticness']
top10_genres = genre_data[genre_data['genres'].isin(top10_genres['genres'])][['genres'] + sound_features]
```

```python
# Tracer le graphique à barres groupées
fig = px.bar(top10_genres, x='genres', y=sound_features, 
             barmode='group', 
             title='Tendance de diverses caractéristiques sonores sur les 10 principaux genres')

fig.update_layout(
    xaxis_title='Genres',
    yaxis_title='Valeur',
    legend_title='Caractéristiques Sonores'
)

fig.show()
```

5) Générer un nuage de mots des genres présents dans l'ensemble de données à l'aide de la bibliothèque WordCloud : WordCloud(width=800, height=800, background_color='white', stopwords=stopwords, max_words=40, min_font_size=10).generate(comment_words)

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Extraire les genres
genres = genre_data['genres'].dropna()

# Combiner tous les genres en une seule chaîne de texte
comment_words = ' '.join(genres)

# Définir les mots à exclure (stopwords)
stopwords = set(STOPWORDS)

# Générer le nuage de mots
wordcloud = WordCloud(width=800, height=800, 
                      background_color='white', 
                      stopwords=stopwords, 
                      max_words=40, 
                      min_font_size=10).generate(comment_words)

# Afficher le nuage de mots
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

9) Identifiez les 10 meilleurs artistes avec le plus de chansons produites et affichez le nombre et le nom de l'artiste : top10_most_song_produced_artists[['count','artists']].sort_values('count', ascending=False).

```python
# Compter le nombre de chansons par artiste
artist_song_count = genre_data['artists'].value_counts().reset_index()

# Renommer les colonnes pour plus de clarté
artist_song_count.columns = ['artists', 'count']

# Sélectionner les 10 artistes avec le plus de chansons produites
top10_most_song_produced_artists = artist_song_count.nlargest(10, 'count')

# Afficher le nombre de chansons et le nom de l'artiste
print(top10_most_song_produced_artists[['count', 'artists']].sort_values('count', ascending=False))
```
